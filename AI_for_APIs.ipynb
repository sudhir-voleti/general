## Third prompt featuring a variation (P3)
Nice, it works. Now, say I have a list of major cities in the world {Tokyo, Beijing, Delhi, Moscow, Paris, New York}.
I'd like to collect current weather data on each and put into a dataframe. Modify code to enable this pls.


```python
import requests
import pandas as pd

# Replace with your actual API key
API_KEY = "13fc2c526a23adc910da505200aa84be"

# List of cities
cities = ["Tokyo", "Beijing", "Delhi", "Moscow", "Paris", "New York"]

# City coordinates (approximate latitude and longitude)
city_coordinates = {
    "Tokyo": {"lat": 35.6895, "lon": 139.6917},
    "Beijing": {"lat": 39.9042, "lon": 116.4074},
    "Delhi": {"lat": 28.7041, "lon": 77.1025},
    "Moscow": {"lat": 55.7558, "lon": 37.6173},
    "Paris": {"lat": 48.8566, "lon": 2.3522},
    "New York": {"lat": 40.7128, "lon": -74.0060}
}

# Base URL for Current Weather API
BASE_URL = "http://api.openweathermap.org/data/2.5/weather"

# List to store weather data for all cities
all_cities_weather_data = []

# Loop through each city
for city_name in cities:
    lat = city_coordinates[city_name]["lat"]
    lon = city_coordinates[city_name]["lon"]

    # Construct the API query URL
    query_url = f"{BASE_URL}?lat={lat}&lon={lon}&appid={API_KEY}"

    # Make the API request
    response = requests.get(query_url)

    # Check if the request was successful
    if response.status_code == 200:
        data = response.json()

        # Extract relevant data
        current_weather_data = {
            'City': city_name,
            'Latitude': data['coord']['lat'],
            'Longitude': data['coord']['lon'],
            'Timestamp': data['dt'],  # Unix timestamp
            'DateTime': pd.to_datetime(data['dt'], unit='s'),  # Readable datetime
            'Temperature (°C)': data['main']['temp'] - 273.15, # Convert from Kelvin to Celsius
            'Description': data['weather'][0]['description'],
            'Wind Speed (m/s)': data['wind']['speed'],
            'Humidity (%)': data['main']['humidity'],
            'Pressure (hPa)': data['main']['pressure'],
            'Cloudiness (%)': data['clouds']['all'] if 'clouds' in data else None, # Handle cases where cloud data might be missing
            'Sunrise': pd.to_datetime(data['sys']['sunrise'], unit='s'),
            'Sunset': pd.to_datetime(data['sys']['sunset'], unit='s')
        }
        all_cities_weather_data.append(current_weather_data)

    else:
        print(f"Error for {city_name}: API request failed with status code {response.status_code}")
        print(response.text) # Print the error message for debugging

# Create a Pandas DataFrame from the collected data
df_all_cities_weather = pd.DataFrame(all_cities_weather_data)

# Print the DataFrame (optional: select and order columns for neatness)
print(df_all_cities_weather[['City', 'DateTime', 'Temperature (°C)', 'Description', 'Wind Speed (m/s)', 'Humidity (%)', 'Pressure (hPa)', 'Cloudiness (%)', 'Sunrise', 'Sunset']].to_string())
```

### Downstream processing prompt (P4)

Attached is a picture of the output dataframe from the API.

In the dataframe above, write me python code to plot and display temperature and windspeed for the different cities.

X-axis should be cities (going east to west). On Y-axis, make a line plot for each of Temp and windspeed (in different colors).

Make it interactive with plotly, so that I can get the values when cursor points to a plot curve.

Make the plot aesthetically appealing (fonts, background colors, labels and legends etc.)

Here's the code output Gemini gave me.

```python
import pandas as pd
import plotly.graph_objects as go
from plotly.subplots import make_subplots
from io import StringIO

# Dataframe from the image (manually transcribed for demonstration)
csv_data = """
,City,DateTime,Temperature (°C),Description,Wind Speed (m/s),Humidity (%),
0,Tokyo,2025-03-04 01:11:08,4.12,moderate rain,7.72,59,
1,Beijing,2025-03-04 01:15:16,2.92,scattered clouds,1.81,28,
2,Delhi,2025-03-04 01:15:16,17.06,mist,2.06,77,
3,Moscow,2025-03-04 01:15:17,1.12,light snow,4.28,77,
4,Paris,2025-03-04 01:15:17,4.47,clear sky,1.54,79,
5,New York,2025-03-04 01:12:11,1.42,clear sky,5.14,54,
"""
df_all_cities_weather = pd.read_csv(StringIO(csv_data), index_col=0)

# Order cities from East to West (approximately based on longitude)
city_order = ["Tokyo", "Beijing", "Delhi", "Moscow", "Paris", "New York"]
df_all_cities_weather['City'] = pd.Categorical(df_all_cities_weather['City'], categories=city_order, ordered=True)
df_all_cities_weather = df_all_cities_weather.sort_values('City')


# Create the Plotly plot
fig = make_subplots(specs=[[{"secondary_y": True}]])

# Add traces for Temperature and Wind Speed
fig.add_trace(go.Scatter(x=df_all_cities_weather['City'],
                         y=df_all_cities_weather['Temperature (°C)'],
                         mode='lines+markers',
                         name='Temperature (°C)',
                         marker=dict(size=8),
                         line=dict(color='firebrick', width=2),
                         hovertemplate="City: %{x}<br>Temperature: %{y}°C"),
              secondary_y=False)

fig.add_trace(go.Scatter(x=df_all_cities_weather['City'],
                         y=df_all_cities_weather['Wind Speed (m/s)'],
                         mode='lines+markers',
                         name='Wind Speed (m/s)',
                         marker=dict(size=8),
                         line=dict(color='royalblue', width=2),
                         hovertemplate="City: %{x}<br>Wind Speed: %{y} m/s"),
              secondary_y=True)

# Update layout for aesthetics
fig.update_layout(
    title={
        'text': '<b>Current Weather: Temperature and Wind Speed Across Cities</b>',
        'x':0.5,
        'xanchor': 'center',
        'font': dict(size=24, color='black')
    },
    xaxis=dict(
        title='<b>City (East to West)</b>',
        titlefont=dict(size=16, color='black'),
        tickfont=dict(size=12, color='black'),
        showline=True,
        linewidth=2,
        linecolor='black',
        mirror=True
    ),
    yaxis=dict(
        title='<b>Temperature (°C)</b>',
        titlefont=dict(size=16, color='firebrick'),
        tickfont=dict(size=12, color='firebrick'),
        showline=True,
        linewidth=2,
        linecolor='black',
        mirror=True,
        gridcolor='lightgray'
    ),
    yaxis2=dict(
        title='<b>Wind Speed (m/s)</b>',
        titlefont=dict(size=16, color='royalblue'),
        tickfont=dict(size=12, color='royalblue'),
        showline=True,
        linewidth=2,
        linecolor='black',
        mirror=True,
        gridcolor='lightgray',
        overlaying='y',
        side='right'
    ),
    plot_bgcolor='rgba(240, 240, 240, 0.8)',  # Light gray background
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=1.02,
        xanchor="right",
        x=1,
        font=dict(size=12, color='black')
    ),
    hovermode="closest" # or 'x unified' for different hover modes
)

# Show the plot
fig.show()
```
